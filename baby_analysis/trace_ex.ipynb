```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
path = "C:/Users/bengieru/Desktop/calcium_analysis/calcium_movies/Output/Output"

trace=open( path +"/m_1005_session_1_baseline_trace.csv", "r")
spike=open( path +"/m_1005_session_1_baseline_spikes_complexity.csv", "r")
feature=open( path +"/m_1005_session_1_baseline_features.csv", "r")
reader_trace = csv.reader(trace)
reader_spike = csv.reader(spike)
reader_feature = csv.reader(feature)
traces=np.zeros((134,8000))
spikes=np.zeros((134,8000))
#row_count = sum(1 for row in reader)
for i,line in enumerate(reader_trace):
    for j in range(len(line)):
        traces[i][j]=line[j]
for i,line in enumerate(reader_spike):
    for j in range(len(line)):
        spikes[i][j]=line[j]
```

```python
fig,ax=plt.subplots(10,1,figsize=(40,175))
for i in range(10):
    y=np.arange(8000)
    ax[i].plot(traces[i],label='traces')
    ax[i].scatter(y,spikes[i],label='spike_max')
    #ax.plot(spikes[0],label='spikes')
    ax[i].legend()
print(len(spikes[1]))
print(len(y))
```
